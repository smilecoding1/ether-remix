```python

def func(x):
    return (x + 3)**2

def grad(x):
    return 2 * (x + 3)


def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    for i in range(num_iterations):
        gradient = grad(x)
        x = x - learning_rate * gradient
        print(f"Iteration {i+1}: x = {x}, f(x) = {func(x)}")
    return x

```

```python

import numpy as np
import matplotlib.pyplot as plt
```

```python

def func(x):
    return (x + 3)**2


def grad(x):
    return 2 * (x + 3)

```

```python

def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    x_history = [x]
    for i in range(num_iterations):
        gradient = grad(x)
        x = x - learning_rate * gradient
        x_history.append(x)
    return x, x_history

```

```python

initial_x = 2
learning_rate = 0.1
num_iterations = 100
```

```python

minimum_x, x_history = gradient_descent(initial_x, learning_rate, num_iterations)
```

```python

x_values = np.linspace(-10, 5, 400)
y_values = func(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='y = (x + 3)^2')
plt.scatter(x_history, func(np.array(x_history)), color='red', label='Gradient Descent Steps')
plt.plot(x_history, func(np.array(x_history)), linestyle='--', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()

print(f"Local minimum occurs at: x = {minimum_x}")
```
